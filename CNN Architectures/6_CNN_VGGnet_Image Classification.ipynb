# Create VGG-16 Model

The code you provided is building a custom implementation of the VGG-16 architecture using the Keras and then making predictions on four different images. It also visualizes the images and their top three class predictions with confidence probabilities.

This section creates the VGG-16 architecture using Keras. The architecture has:

Convolutional layers (Conv2D) with ReLU activation: **13 convolutional layers** are used, organized into five blocks.

Max-pooling layers (MaxPooling2D) to reduce the spatial dimensions.

Dense (fully connected) layers at the end, with the final layer having 1,000 units (corresponding to 1,000 classes in the ImageNet dataset) and a softmax activation function.

Block 1: Convolution + Max Pooling

Block 2: Convolution + Max Pooling

 Block 3: Convolution + Max Pooling

 Block 4: Convolution + Max Pooling

 Block 5: Convolution + Max Pooling

Flatten Layer

**Fully Connected Layers:** there are two fully connected layers of size 4096.
The final output layer is a 1,000-class softmax layer, typically used for ImageNet classification.

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model

_input = Input((224,224,1))

conv1  = Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu")(_input)
conv2  = Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu")(conv1)
pool1  = MaxPooling2D((2, 2))(conv2)

conv3  = Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu")(pool1)
conv4  = Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu")(conv3)
pool2  = MaxPooling2D((2, 2))(conv4)

conv5  = Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(pool2)
conv6  = Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(conv5)
conv7  = Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(conv6)
pool3  = MaxPooling2D((2, 2))(conv7)

conv8  = Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(pool3)
conv9  = Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv8)
conv10 = Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv9)
pool4  = MaxPooling2D((2, 2))(conv10)

conv11 = Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(pool4)
conv12 = Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv11)
conv13 = Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(conv12)
pool5  = MaxPooling2D((2, 2))(conv13)

flat   = Flatten()(pool5)
dense1 = Dense(4096, activation="relu")(flat)
dense2 = Dense(4096, activation="relu")(dense1)
output = Dense(1000, activation="softmax")(dense2)

vgg16_model  = Model(inputs=_input, outputs=output)
```

```python
# Save the model to a file
model_path = '/content/vgg_16_model.h5'
vgg16_model.save(model_path)
```

The first part of _get_predictions() displays the images in a 1x4 grid using Matplotlib.
The images are resized to 200x200 pixels for visualization purposes.

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns
import numpy as np

# Define image paths
img1 = "/content/carw.jpg"
img2 = "/content/panda.jpg"
img3 = "/content/tomato.jpg"
img4 = "/content/ladyfinger.jpg"
imgs = [img1, img2, img3, img4]


def _load_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the image
    return img

def _get_predictions(_model):
    # Display images
    f, ax = plt.subplots(1, 4)
    f.set_size_inches(80, 40)
    for i in range(4):
        ax[i].imshow(Image.open(imgs[i]).resize((200, 200), Image.LANCZOS))
        ax[i].axis('off')  # Hide axes
    plt.show()

    # Display predictions
    f, axes = plt.subplots(1, 4)
    f.set_size_inches(80, 20)
    for i, img_path in enumerate(imgs):
        img = _load_image(img_path)
        preds = _model.predict(img)
        top_preds = np.argsort(preds[0])[-3:][::-1]  # Get indices of top 3 predictions
        top_probs = preds[0][top_preds]
        top_labels = [f'Class {idx}' for idx in top_preds]  # Replace with actual class names if available
        b = sns.barplot(y=top_labels, x=top_probs, color="gray", ax=axes[i])
        b.tick_params(labelsize=20)
        axes[i].set_title(f'Image {i+1}', fontsize=20)
        axes[i].set_xlabel('Probability', fontsize=20)
        axes[i].set_ylabel('Class Label', fontsize=20)
    plt.tight_layout()
    plt.show()

# Get predictions and display results
_get_predictions(model)
```
